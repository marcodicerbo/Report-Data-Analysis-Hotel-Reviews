### Origine e descrizione del dataset

#### Acknowledgements

I dati sono stati estratti da Booking.com. Tutti i dati contenuti nel file sono già pubblicamente disponibili. I dati sono di proprietà di Booking.com.

#### Contenuto

Il dataset contiene circa 515.000 recensioni e valutazioni dei clienti relative a 1493 hotel di lusso in tutta Europa. Sono presenti 17 colonne, descritte di seguito:
- Hotel_Address: Indirizzo dell'hotel;

- Review_Date: Data in cui il cliente ha pubblicato la recensione corrispondente;

- Average_Score: Punteggio medio dell'hotel;

- Hotel_Name: Nome dell'hotel;

- Reviewer_Nationality: Nazionalità del cliente;

- Negative_Review: Recensione negativa;

- Review_Total_Negative_Word_Counts: Numero totale di parole nella recensione negativa;

- Positive_Review: Recensione positiva;

- Review_Total_Positive_Word_Counts: Numero totale di parole nella recensione positiva;

- Reviewer_Score: Punteggio assegnato dal cliente all'hotel;

- Total_Number_Of_Reviews_Reviewer_Has_Given (in seguito rinominato in Reviewer_Review_Count): Numero di recensioni scritte dal cliente in passato;

- Total_Number_Of_Reviews: Numero totale di recensioni dell'hotel;

- Tags: Tag assegnati dal cliente all'hotel;

- days_since_review: Durata tra la data della recensione e la data di acquisizione dei dati;

- Additional_Number_of_Scoring: Alcuni ospiti hanno assegnato un punteggio al servizio senza una recensione. Questo numero indica quanti punteggi senza recensione sono presenti;

- lat: Latitudine dell'hotel;

- lng: Longitudine dell'hotel.

### Data Cleaning

#### Gestione missing values
Dopo aver verificato la presenza di valori mancanti ed aver appurato che sono presenti soltanto nelle colonne lat e lng, procedo alla loro eliminazione in quanto rispetto al totale di oltre 500k righe possono essere trascurati.

```python
# Gestione missing values
dfhrev2.dropna(subset=['lat', 'lng'], inplace=True) # elimino righe con valori nulli nelle colonne lat e lng
dfhrev2.isnull().sum() # verifico di aver eliminato i valori nulli
```

#### Conversione in datetime ed estrazione anno/mese
Avendo osservato che la colonna Review_Date è di tipo object, la converto in datetime per una maggiore accuratezza. Inoltre, per facilitare analisi successive creo delle nuove colonne dove inserire esclusivamente anno, mese ed anno e mese insieme, quest'ultima utile per poter raggruppare le recensioni in base ad un periodo mensile.

```python
# Conversione in datetime
dfhrev2['Review_Date'] = pd.to_datetime(dfhrev2['Review_Date']) # converto la colonna Review_Date da tipo object a datetime
print(dfhrev2['Review_Date'].dtype) # controllo che il dato sia stato convertito correttamente
```

```python
# Estrazione anno e mese creando nuove colonne
dfhrev2['Review_Year'] = dfhrev2['Review_Date'].dt.year # estraggo l'anno 
dfhrev2['Review_Month'] = dfhrev2['Review_Date'].dt.month # estraggo il mese
dfhrev2['Review_YearMonth'] = dfhrev2['Review_Date'].dt.to_period('M') # estraggo anno e mese insieme creando un periodo temporale mensile, dove 'M' = Monthly (mese)
print("Estrazione anno e mese:") # controllo le nuove colonne
print(dfhrev2[['Review_Date', 'Review_Year', 'Review_Month', 'Review_YearMonth']].head())
```

#### Creazione variabili derivate
Per definire la lunghezza totale di ciascuna recensione, creo una nuova colonna, chiamata Review_Length, in cui sommare le parole positive e negative presenti nelle recensioni. Inoltre, per maggiore chiarezza e comodità, rinomino la colonna Total_Number_of_Reviews_Reviewer_Has_Given in Reviewer_Review_Count. 

```python
# Creazione variabili derivate
dfhrev2['Review_Length'] = (dfhrev2['Review_Total_Negative_Word_Counts'] + 
                                 dfhrev2['Review_Total_Positive_Word_Counts']) # creo una nuova colonna per definire la lunghezza totale della recensione
dfhrev2['Reviewer_Review_Count'] = dfhrev2['Total_Number_of_Reviews_Reviewer_Has_Given'] # rinomino la colonna 

```

#### Gestione anomalie nomi hotel ed indirizzi
L'analisi preliminare del dataset ha evidenziato due anomalie strutturali nella corrispondenza tra i nomi degli hotel e gli indirizzi.

##### Anomalia 1: Indirizzo con più nomi (1 caso)
Indirizzo: 
- "8 Northumberland Avenue Westminster Borough London WC2N 5BY"

Nomi: 
- "The Grand at Trafalgar Square"
- "Club Quarters Hotel Trafalgar Square"
Interpretazione: Probabile rebranding o gestione separata su Booking.com dello stesso immobile fisico.

Soluzione: Gli indirizzi sono stati gestiti come 2 record distinti.
Risultato: 1476 combinazioni univoche (invece di 1475 attese).

##### Anomalia 2: Nome con più indirizzi (1 caso)
Hotel Regina appare in:
- Bergara 2-4, Barcellona (Spagna)
- Rooseveltplatz 15, Vienna (Austria)
- Via Cesare Correnti 13, Milano (Italia)
Interpretazione: Catena alberghiera; "Hotel Regina" con sedi multiple in Europa.
Soluzione: Gli hotel sono stati getsiti come 3 record distinti.
Risultato: 1.477 Hotel distinti (invece di 1.475 attesi).

##### Resoconto finale
Al fine di utilizzare il DataFrame per successive analisi in SQL e la crezione di identificativi per ciascun hotel (hotel_id), si è scelto di trattare gli hotel con anomalie come entità distinte: 
1.475 hotel (1 nome = 1 indirizzo)
1 hotel multi-nome (1 indirizzo = 2 nomi)
1 catena multi-sede (1 nome = 3 indirizzi)
Risultato: 1.477 hotel_id univoci, tutte le recensioni preservate.

##### Creazione tabella reviews

```python
# Ripulisco l'indice: da 0 a N-1, senza "buchi" da operazioni precedenti
dfhrev2 = dfhrev2.reset_index(drop=True)

# Creo la chiave surrogata review_id come indice + 1 (parte da 1, non da 0)
dfhrev2["review_id"] = dfhrev2.index + 1

# Verifico che review_id sia effettivamente univoco
print("review_id univoco:", dfhrev2["review_id"].is_unique)
print("Numero di recensioni:", dfhrev2.shape[0])
```

Ho creato una chiave surrogata review_id assegnando un ID progressivo a ogni riga del dataset pulito. Questa chiave funge da primary key nella tabella reviews e garantisce un identificativo univoco per ciascuna recensione.

```python
# merge con hotels per ricavare hotel_id 
reviews_with_hotels = dfhrev2.merge(
    hotels_df[["hotel_id", "hotel_name", "hotel_address"]],
    left_on=["Hotel_Name", "Hotel_Address"],
    right_on=["hotel_name", "hotel_address"],
    how="left"  # LEFT: mantengo tutte le recensioni, anche se qualche hotel non matcha
)

# Controllo quante recensioni non hanno trovato un hotel_id
missing_hotels = reviews_with_hotels["hotel_id"].isna().sum()
print("Recensioni senza hotel_id dopo la merge:", missing_hotels)
```

Per collegare ogni recensione al relativo hotel, ho effettuato una join tra il dataset in cui sono presenti le recensioni e hotels, utilizzando il nome e l’indirizzo dell’hotel. In questo modo ho recuperato l'hotel_id già definito nella tabella hotels e ho verificato il numero di recensioni per cui l’identificativo non è stato trovato.

```python
# merge con reviewers per ricavare reviewer_id
reviews_with_ids = reviews_with_hotels.merge(
    reviewers_df[["reviewer_id", "reviewer_nationality", "reviewer_review_count"]],
    left_on=["Reviewer_Nationality", "Reviewer_Review_Count"],
    right_on=["reviewer_nationality", "reviewer_review_count"],
    how="left"  # LEFT: mantengo tutte le recensioni
)

# Controllo quante recensioni non hanno trovato un reviewer_id
missing_reviewers = reviews_with_ids["reviewer_id"].isna().sum()
print("Recensioni senza reviewer_id dopo la merge:", missing_reviewers)
```

In modo analogo, ho unito il dataset delle recensioni con reviewers sfruttando la coppia (Reviewer_Nationality, Reviewer_Review_Count) come chiave di collegamento. Questa join mi ha permesso di associare a ogni recensione il reviewer_id già presente nella tabella dei reviewers.

```python
reviews_df = reviews_with_ids[[
    "review_id",
    "hotel_id",
    "reviewer_id",
    "Review_Date",
    "Review_Year",
    "Review_Month",
    "Review_YearMonth",
    "Review_Length",
    "days_since_review",
    "Negative_Review",
    "Positive_Review",
    "Review_Total_Positive_Word_Counts",
    "Review_Total_Negative_Word_Counts",
    "Reviewer_Score"
]].rename(columns={
    "Review_Date": "review_date",
    "Review_Year": "review_year",
    "Review_Month": "review_month",
    "Review_YearMonth": "review_yearmonth",
    "Review_Length": "review_length",
    "Negative_Review": "negative_review",
    "Positive_Review": "positive_review",
    "Review_Total_Positive_Word_Counts": "review_total_positive_word_counts",
    "Review_Total_Negative_Word_Counts": "review_total_negative_word_counts",
    "Reviewer_Score": "reviewer_score"
})

print("Shape finale reviews_df:", reviews_df.shape)
print(reviews_df.head())
```

A partire dal dataset arricchito con le chiavi esterne, ho costruito il dataframe finale reviews_df selezionando esclusivamente gli attributi necessari per la tabella reviews e rinominando le colonne in quanto in sql i nomi sono riportati tutti in minuscolo. Il risultato è una tabella con una riga per recensione, primary key review_id e foreign key hotel_id e reviewer_id.
